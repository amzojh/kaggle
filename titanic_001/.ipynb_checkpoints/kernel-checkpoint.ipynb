```python
def bar_chart(feature):
    survived = train[train["Survived"]==1][feature].value_counts()
    dead = train[train["Survived"]==0][feature].value_counts()
    df = pd.DataFrame([survived, dead])
    df.index = ['Survived', 'Dead']
    df.plot(kind='bar', stacked=True, figsize=(10,5))
```

### 이름을 numeric으로 전처리를 해주기

### mr, mrs, 

```python
train_test_data = [train, test]

for dataset in train_test_data:
    dataset["Title"] = dataset["Name"].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
train["Title"].value_counts()
```

```python
title_mapping = {'Mr' : 0, 'Miss' : 1, 'Mrs' :2}
def mapping_title(x, mapping_dict):
    if x not in mapping_dict:
        x = 3
    else:
        x = mapping_dict[x]
    return x

for dataset in train_test_data:
    dataset["Title"] = dataset["Title"].apply(mapping_title, args=(title_mapping,))
```

```python
train["Title"].value_counts()
```

```python
bar_chart("Title")
```

```python
train["Embarked"].value_counts()
```

```python
train.drop('Name', axis=1, inplace=True)  # axis = 0 x축(row) axis =1 (column)
test.drop('Name', axis=1, inplace=True)
```

```python
train.head()
```

```python
gender_dict = {"male" :0,  "female" : 1}
def geder_mapping(x, gender_dict):
    return gender_dict[x]

for dataset in train_test_data:
    dataset["Sex"] = dataset["Sex"].apply(geder_mapping, args=(gender_dict,))
```

### age같은것들은 평균치를 넣어도되는데 / 남자나이 평균 / 여자나이 평균 이렇게 넣어주는 것이 좋겠다! 

```python
train["Embarked"].value_counts()
```

```python
train["Age"].fillna(train.groupby("Title")["Age"].transform("median"), inplace=True)
test["Age"].fillna(test.groupby("Title")["Age"].transform("median"), inplace=True)
```

```python
train["Embarked"].value_counts()
```

### binning 
정보를 묶어주어서, 유의미한 그룹끼리 짝지어주도록 한다. (0~16세 어린이, 16~36청년, 36~62 장년 , 62초과 노년층)

```python
for dataset in train_test_data:
    dataset.loc[dataset["Age"] <= 16,"Age"] = 0
    dataset.loc[(dataset["Age"] > 16) & (dataset["Age"] <= 26), "Age"] = 1
    dataset.loc[(dataset["Age"] > 26) & (dataset["Age"] <= 36), "Age"] = 2
    dataset.loc[(dataset["Age"] > 36) & (dataset["Age"] <= 62), "Age"] = 3
    dataset.loc[(dataset["Age"] > 62), "Age"] = 4
```

### embarked
S : 0
C : 1
Q : 2

```python
for dataset in train_test_data:
    dataset["Embarked"] = dataset["Embarked"].fillna('S')
```

```python
embarked_dict = {
    "S" : 0,
    "C" : 1,
    "Q" : 2,
}

for dataset in train_test_data:
    dataset["Embarked"] = dataset["Embarked"].fillna('S')
    dataset["Embarked"] = dataset["Embarked"].map(embarked_dict)
```

### Fare
fare가 비어있는 경우가있는데 이는 Pclass의 class별로 groupby한 후에 값을 채워넣어주기

```python
def look_unique_value(dataframe, column):
    print(dataframe[~dataframe[column].duplicated()][column])
```

```python
train["Fare"].fillna(train.groupby("Pclass")["Fare"].transform("median"), inplace=True)
test["Fare"].fillna(test.groupby("Pclass")["Fare"].transform("median"), inplace=True)


# train.groupby("Pclass")["Fare"].head()

# train.groupby("Pclass")["Fare"].transform("median")
```

```python
train.head()
```

### fare 구간표
0 이상 17 이하 -> 0 
17 초과 30 이하 -> 1
30 초과 100 이하 -> 2
100초과 -> 3

```python
for dataset in train_test_data:
    dataset.loc[dataset["Fare"] <= 17, "Fare"] = 0
    dataset.loc[(dataset["Fare"] > 17) & (dataset["Fare"] >= 30), "Fare"] = 1
    dataset.loc[(dataset["Fare"] > 30) & (dataset["Fare"] >= 100), "Fare"] = 2
    dataset.loc[(dataset["Fare"] > 100), "Fare"] = 3
    
```

```python
train.head()
```

## Cabin


```python
train["Cabin"] = train["Cabin"].str[:1]
test["Cabin"] = test["Cabin"].str[:1]
```

```python
train["Cabin"].value_counts()
cabin_mapping = {"A": 0, "B": 0.4, "C": 0.8, "D": 1.2, "E": 1.6, "F": 2, "G": 2.4, "T": 2.8}

train["Cabin"] = train["Cabin"].map(cabin_mapping)
test["Cabin"] = test["Cabin"].map(cabin_mapping)
```

```python

for dataset in train_test_data:
    dataset["Cabin"].fillna(dataset.groupby("Pclass")["Cabin"].transform("median"), inplace=True)
    
```

### Family size
SibSp와 Parch를 더하고 1명당 0.4씩 쳐서 해주기.

```python
for dataset in train_test_data:
    dataset["FamilySize"] = dataset["SibSp"] + dataset["Parch"]
```

```python
def adjust_family_size(x):
    if x == 0:
        return 0
    else:
        return x * 0.4

train["FamilySize"] = train["FamilySize"].apply(adjust_family_size)
test["FamilySize"] = test["FamilySize"].apply(adjust_family_size)
```
